```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import pandas_datareader.data as web
```

```python
df = web.DataReader("AAPL", "yahoo", start="2012-9-1", end="2017-8-31")
```

```python
df.head()
```

```python
P = df["Adj Close"]
```

```python
P.plot()
```

```python
r = P.diff()/P
```

```python
r[-100:].plot()
```
