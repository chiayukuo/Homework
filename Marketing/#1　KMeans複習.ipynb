### 匯入CSV

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Consumer_data.csv')
```

```python
dataset
```

```python
x=dataset.iloc[:,[2,3]].values  #只看年齡、收入的部分
```

```python
x
```

```python
plt.scatter(x=x[:,0],y=x[:,1],color='blue')  
plt.xlabel('Age')
plt.ylabel('Income')

plt.show()
```

### KMeans

```python
from sklearn.cluster import KMeans
```

```python
clf = KMeans(n_clusters=3)
```

```python
clf.fit(x)
```

```python
clf.labels_
```

```python
plt.scatter(x[:,0],x[:,1], c=clf.labels_)
```

```python
dataset['label'] = clf.labels_
```
