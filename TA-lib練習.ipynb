```python
import yfinance as yf

# 取得 AAPL 價量資料
df = yf.Ticker('AAPL').history(period = 'max')
```

```python
df
```

# 資料需求：

* high：最高價
* low：最低價
* close：收盤價

# 參數設定：
* timeperiod：天數參數，預設為 14

```python
talib.ADX(df.High, df.Low, df.Close, timeperiod = 14)
```

```python
talib.RSI(df.Close)
```
