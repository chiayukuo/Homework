# 超簡易預測大立光股價
* 參考：IT邦幫忙

```python
LARGAN = tejapi.get('TRAIL/TAPRCD', coid='3008')
```

```python
LARGAN
```

```python
LARGAN = LARGAN.set_index('mdate')
```

```python
LARGAN
```

```python
import numpy as np
import pandas as pd
```

```python
import datetime as datetime

from fbprophet import Prophet
```

```python
#繪製收盤價趨勢圖

LARGAN['close_d'].plot()
```

```python
from sklearn import metrics
```

```python
new_LARGAN = pd.DataFrame(LARGAN['close_d']).reset_index().rename(columns={'mdate':'ds', 'close_d':'y'})
```

```python
new_LARGAN
```

```python
new_LARGAN['ds'][0] = new_LARGAN['ds'][0].replace(tzinfo=None)
```

### TEJAPI的時間資料需要額外進行處理：把timezone移除、datatime切換等等……
### 還是twstock比較方便嗎！！？

```python
for i in range(242):
    new_LARGAN['ds'][i] = new_LARGAN['ds'][i].replace(tzinfo=None)
```

```python
for j in range(242):
    new_LARGAN['ds'][j] =  new_LARGAN['ds'][j].strftime('%Y-%m-%d')
```

```python
new_LARGAN['y'] = np.log(new_LARGAN['y'])

# 定義模型

model = Prophet()

model.fit(new_LARGAN)

# 建構預測集

future = model.make_future_dataframe(periods=242)

# 進行預測

forecast = model.predict(future)

forecast.head()
```

```python
figure=model.plot(forecast)   #藍色區域代表誤差範圍
```
