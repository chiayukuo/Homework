### 擷取台積電、大日光、元大台灣50反1自武漢肺炎爆發後之股價

```python
import pandas_datareader as pdr
import datetime
```

```python
start = datetime.datetime(2019,12,19)
df_2330 = pdr.DataReader('2330.TW', 'yahoo', start=start)
df_3008 = pdr.DataReader('3008.TW', 'yahoo', start=start)
df_00632R = pdr.DataReader('00632R.TW','yahoo',start=start)
```

```python
for stock in [df_2330, df_3008 ,df_00632R]:
    stock['normalized_price']=stock['Adj Close']/stock['Adj Close'].iloc[0]    #除以買入時點價格
```

```python
df_2330
```

### 計算收益率

```python
for stock, weight in zip([df_2330,df_3008,df_00632R],[0.4, 0.4, 0.2]):
    stock['weighted daily return'] = stock['normalized_price']*weight
```

```python
import pandas as pd
```

```python
df_total=pd.concat([df_2330['weighted daily return'] , df_3008['weighted daily return'],df_00632R['weighted daily return']],axis=1)
df_total.columns=['2330', '3008','00632R']     #將兩表格合併
```

```python
df_total_money = df_total* 100000
df_total_money['Total Pos']=df_total_money.sum(axis=1)  #加總方向
df_total_money.head()
```

### 收益曲線繪製

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.plot(df_total_money['2330'])
plt.plot(df_total_money['3008'])
plt.plot(df_total_money['00632R'])

plt.grid(True, axis='y')

plt.legend()
```
