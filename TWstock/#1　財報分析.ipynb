```python
import requests
```

```python
from selenium import webdriver
from pyquery import PyQuery as pq
```

```python
driver = webdriver.Chrome('chromedriver.exe')
driver.get('https://mops.twse.com.tw/mops/web/t163sb06')
```

```python
html = driver.find_element_by_css_selector("*").get_attribute("outerHTML")
doc = pq(html)
```

```python
doc
```

### 送出表單

```python
driver.find_element_by_css_selector("#search_bar1 > div > input[type=button]").click()
```

```python
html = driver.find_element_by_css_selector("*").get_attribute("outerHTML")
doc = pq(html)
```

```python
info=doc('.odd')
```

```python
info
```

```python
stock = []

for data in info:
        stock.append(doc(data).text())
```

```python
stock
```

## 組成DataFrame

```python
td = []
for i in range(len(stock)):
   td.append(stock[i].split('\n'))
```

```python
td
```

```python
import pandas as pd
df = pd.DataFrame(data=td)
```

```python
df
```

```python
df.columns=['公司代號','公司名稱','營業收入(百萬元)','毛利率(%)','營業利益率(%)','稅前純益率(%)','稅後純益率(%)']
```

```python
df.index = df['公司名稱'] #更改index
df = df.drop(columns = '公司名稱' )
```
