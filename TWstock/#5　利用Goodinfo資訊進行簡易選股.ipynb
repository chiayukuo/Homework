### 匯入CSV

```python
import pandas as pd
```

```python
df = pd.read_csv("StockList (1).csv", thousands=",")   #去除千分位的逗號
```

```python
df
```

```python
df.index = df['名稱'] #更改index
df = df.drop(columns = '名稱' )
df = df.drop(columns = '一個月走勢圖' )
df = df.drop(columns = '三個月走勢圖' )
```

```python
df
```


### 篩出條件符合的股票

```python
cond1 = df['風險係數(β)'] < 1.4
cond2 = df['累季ROA(%)'] > 10
cond3 = df['累季營收年增(%)'] > 9
cond4 = df['PER'] < 18
cond5 = df['歷年加均ROA(%)'] > 10
cond6 = df['歷年加均毛利成長(%)'] > 5
cond7 = df['近4Q營益季增(%)'] > 3

df_result = df[cond1 & cond2 & cond3 & cond4 & cond5 & cond6 & cond7]
```

```python
df_result.sort_values(by="風險係數(β)",ascending= True)
```

### 改用yahoo finance爬爬看～希望不會被擋

```python
import pandas_datareader as pdr
import datetime
```

```python
start = datetime.datetime(2017,12,19)
df_3533 = pdr.DataReader('3533.TW', 'yahoo', start=start)
df_8464 = pdr.DataReader('8464.TW', 'yahoo', start=start)
df_8499 = pdr.DataReader('8499.TW', 'yahoo', start=start)
df_9910 = pdr.DataReader('9910.TW', 'yahoo', start=start)
df_8437 = pdr.DataReader('8437.TWO', 'yahoo', start=start)
df_8480 = pdr.DataReader('8480.TW', 'yahoo', start=start)
df_4107 = pdr.DataReader('4107.TWO', 'yahoo', start=start)
df_6472 = pdr.DataReader('6472.TWO', 'yahoo', start=start)
df_3541 = pdr.DataReader('3541.TWO', 'yahoo', start=start)
df_4163 = pdr.DataReader('4163.TWO', 'yahoo', start=start)
```

```python
df_3533
```

```python
df_yahoo = pd.concat([df_3533['Close'],df_8464['Close'],df_8499['Close'],df_3541['Close'],df_4107['Close'],
                     df_4163['Close'],df_6472['Close'],df_8437['Close'],df_8480['Close'],df_9910['Close']], axis = 1)
```

```python
df_yahoo.columns=['3533','8464','8499','3541','4107','4163','6472','8437','8480','9910']
```

```python
df_yahoo
```

```python
corr = df_yahoo.corr(method = 'pearson', min_periods = 1)#pearson方法计算相关性
print(corr)
```
