# INSERT 操作

```python
import requests
from bs4 import BeautifulSoup
import sqlite3

#資料庫位置
conn = sqlite3.connect('test.db') # 代表路徑
c = conn.cursor()

url = 'https://tw.buy.yahoo.com/activity/activity950?p=act2-z26-b-180320-channelweather'

res = requests.get(url)
soup = BeautifulSoup(res.text, 'html.parser')


for asus in soup.find_all(class_='C329B1_img_frame'):
    values = list(asus.stripped_strings)

    if '補貨中' not in values:
        values.pop(1)
        values.pop(1)
        values.pop(1)
        c.execute('insert into test(title, price) values (?, ?)', values)
        conn.commit()
        
        print(values)
```

# 查詢並印出

```python
cursor = conn.execute("SELECT id, price  from test")

for row in cursor:
   print("ID = ", row[0])
   print("PRICE = ", row[1])

print ("Operation done successfully")
conn.close()
```

# UPDATE 操作

```python
import sqlite3

conn = sqlite3.connect('test.db')

conn.execute("UPDATE test set price = 25000 where id=1")
conn.commit()

print("Total number of rows updated :", conn.total_changes)
```

# DELETE 操作

```python
import sqlite3

conn = sqlite3.connect('test.db')

conn.execute("DELETE from test where id=2")
conn.commit()
```
