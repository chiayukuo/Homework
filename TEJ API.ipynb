# 台股大盤加權指數

* 目標：繪製簡易走勢圖

```python
data = tejapi.get('TRAIL/TAPRCD',coid='Y9999')
```

```python
data.head()
```

```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties   #新增中文字體
```

```python

data['open_d'].plot(color='green')    #收盤價

data['close_d'].plot(color='red')    #收盤價

plt.grid(True)

plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei'] 
plt.rcParams['axes.unicode_minus'] = False
plt.title("走勢圖")


plt.show()
```

```python
print("開盤價平均數：",data['open_d'].mean())
print("收盤價平均數：",data['close_d'].mean())
```

```python
print("開盤價變異數：",data['open_d'].var())
print("收盤價變異數：",data['close_d'].var())
```

```python
print("開盤價、收盤價之相關係數：",data.open_d.corr(data.close_d))
```
